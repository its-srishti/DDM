```python
from ddm_core import simulate_ddm, fit_ddm
import numpy as np
```

```python
# Step 1: Simulate data with known parameters using my simulator
true_params = {'v': 0.3, 'a': 1.2, 'z': 0.5, 't0': 0.15}
data = simulate_ddm(**true_params, n_trials=500)
```

```python
# Step 2: Fit with my implementation
your_estimates = fit_ddm(data)
print("Your MLE estimates:")
print(f"  v={your_estimates['v']:.3f}  "
      f"a={your_estimates['a']:.3f}  "
      f"z={your_estimates['z']:.3f}  "
      f"t0={your_estimates['t0']:.3f}")
```

```python
import pyddm
```

```python
# Convert data to PyDDM format
# PyDDM expects a Sample object
rts_upper = data[data[:,1]==1, 0].tolist()
rts_lower = data[data[:,1]==0, 0].tolist()
pyddm_data = np.column_stack([data[:,0], data[:,1].astype(int)])
sample = pyddm.Sample.from_numpy_array(pyddm_data, column_names=[])
```

```python
# Fit using PyDDM's built-in MLE
max_rt = data[:,0].max()
model = pyddm.Model(
    drift=pyddm.DriftConstant(drift=pyddm.Fittable(minval=-3, maxval=3)),
    noise=pyddm.NoiseConstant(noise=1),
    bound=pyddm.BoundConstant(B=pyddm.Fittable(minval=0.1, maxval=3)),
    overlay=pyddm.OverlayNonDecision(
        nondectime=pyddm.Fittable(minval=0, maxval=0.4)
    ),
    dx=0.01, dt=0.01,
    T_dur=max_rt + 0.1  # adding a small buffer above RT for my data
)
pyddm.fit_adjust_model(sample, model)

print("\nPyDDM estimates:")
print(f"  v={model.get_model_parameters()[0]:.3f}  "
      f"a={model.get_model_parameters()[1]:.3f}  "
      f"t0={model.get_model_parameters()[2]:.3f}")
```
