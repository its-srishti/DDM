```python
import sys
sys.path.append(r'C:\Users\ASUS\Documents\DDM')

from ddm_core import simulate_ddm, fit_ddm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Study 2: Varying true drift rate and starting point
v_values = [0.1, 0.3, 0.5, 0.8]
z_values = [0.3, 0.5, 0.7]
sample_sizes = [50, 100, 200, 500]
n_simulations = 500
```

```python
results_extended = []

for true_v in v_values:
    for true_z in z_values:
        for n in sample_sizes:
            for sim in range(n_simulations):
                true_params = {
                    'v': true_v, 'a': 1.2,
                    'z': true_z, 't0': 0.15
                }
                data = simulate_ddm(**true_params, n_trials=n)
                estimated = fit_ddm(data)
                
                results_extended.append({
                    'n_trials': n, 'sim': sim,
                    'true_v': true_v, 'true_z': true_z,
                    'est_v': estimated['v'],
                    'est_a': estimated['a'],
                    'est_z': estimated['z'],
                    'est_t0': estimated['t0']
                })
                
                print(f"v={true_v} z={true_z} n={n} "
                      f"sim={sim+1}/{n_simulations} | "
                      f"v_est={estimated['v']:.3f} "
                      f"z_est={estimated['z']:.3f}")
                
                # Save after every fit
                pd.DataFrame(results_extended).to_csv(
                    'recovery_results_extended.csv', index=False)
'''

Total fits: 4 v values × 3 z values × 4 sample sizes × 500 simulations = 24000 fits.

'''
```

```python
df = pd.read_csv('recovery_results_extended.csv')
v08 = df[df['true_v'] == 0.8]

# Check for outliers
print("v=0.8 outlier estimates (|v_est| > 5):")
outliers = v08[v08['est_v'].abs() > 5]
print(f"Count: {len(outliers)} out of {len(v08)}")
print(f"Max |v_est|: {v08['est_v'].abs().max():.3f}")

# Compare mean with and without outliers
print(f"\nMean v_est (all):    {v08['est_v'].mean():.3f}")
print(f"Mean v_est (no outliers): {v08[v08['est_v'].abs() <= 5]['est_v'].mean():.3f}")
```

```python
# Winsorise outliers — cap at ±5 for v estimates
df_clean = df.copy()
df_clean['est_v'] = df_clean['est_v'].clip(-5, 5)

# Save clean version
df_clean.to_csv('recovery_results_extended_clean.csv', index=False)
print("Cleaned. Outliers winsorised at ±5.")
print(f"Rows affected: {(df['est_v'].abs() > 5).sum()}")
```

```python
df_clean = pd.read_csv('recovery_results_extended_clean.csv')

v_values = [0.1, 0.3, 0.5, 0.8]
z_values = [0.3, 0.5, 0.7]
sample_sizes = [50, 100, 200, 500]
params = ['v', 'a', 'z', 't0']

# ── TABLE: Bias and RMSE by true_v and sample size ──
print("BIAS by true_v and sample size (parameter: v)")
print("=" * 60)
print(f"{'true_v':>8} | {'n=50':>8} {'n=100':>8} {'n=200':>8} {'n=500':>8}")
print("-" * 60)
for tv in v_values:
    row = f"{tv:>8.1f} |"
    for n in sample_sizes:
        sub = df_clean[(df_clean['true_v']==tv) & (df_clean['n_trials']==n)]
        bias = (sub['est_v'] - sub['true_v']).mean()
        row += f" {bias:>8.4f}"
    print(row)

print()
print("RMSE by true_v and sample size (parameter: v)")
print("=" * 60)
print(f"{'true_v':>8} | {'n=50':>8} {'n=100':>8} {'n=200':>8} {'n=500':>8}")
print("-" * 60)
for tv in v_values:
    row = f"{tv:>8.1f} |"
    for n in sample_sizes:
        sub = df_clean[(df_clean['true_v']==tv) & (df_clean['n_trials']==n)]
        rmse = np.sqrt(((sub['est_v'] - sub['true_v'])**2).mean())
        row += f" {rmse:>8.4f}"
    print(row)
```

```python
import matplotlib.pyplot as plt
import numpy as np

df_clean = pd.read_csv('recovery_results_extended_clean.csv')
sample_sizes = [50, 100, 200, 500]
colors = {50:'#e74c3c', 100:'#e67e22', 200:'#27ae60', 500:'#2980b9'}

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# ── Panel 1: drift rate v ──
ax = axes[0]
for n in sample_sizes:
    subset = df_clean[df_clean['n_trials'] == n]
    ax.scatter(subset['true_v'], subset['est_v'],
               alpha=0.1, s=8, color=colors[n])
    for tv in subset['true_v'].unique():
        mean_est = subset[subset['true_v']==tv]['est_v'].mean()
        ax.scatter(tv, mean_est, s=80, color=colors[n],
                   zorder=5, edgecolors='white', linewidths=0.8)

# Diagonal
lims = [-0.1, 1.0]
ax.plot(lims, lims, 'k--', linewidth=1.5, alpha=0.6,
        label='Perfect recovery')
ax.set_xlabel('True v', fontsize=12)
ax.set_ylabel('Estimated v', fontsize=12)
ax.set_title('Drift Rate v — True vs Estimated', fontsize=13, fontweight='bold')
handles = [plt.scatter([],[], color=colors[n], label=f'n={n}', s=30)
           for n in sample_sizes]
handles.append(plt.Line2D([0],[0], color='k', linestyle='--',
               label='Perfect recovery'))
ax.legend(handles=handles, fontsize=9)
ax.grid(True, alpha=0.3)
ax.set_xlim(-0.05, 0.95)

# ── Panel 2: starting point z ──
ax = axes[1]
for n in sample_sizes:
    subset = df_clean[df_clean['n_trials'] == n]
    ax.scatter(subset['true_z'], subset['est_z'],
               alpha=0.1, s=8, color=colors[n])
    for tz in subset['true_z'].unique():
        mean_est = subset[subset['true_z']==tz]['est_z'].mean()
        ax.scatter(tz, mean_est, s=80, color=colors[n],
                   zorder=5, edgecolors='white', linewidths=0.8)

lims = [0.1, 0.9]
ax.plot(lims, lims, 'k--', linewidth=1.5, alpha=0.6)
ax.set_xlabel('True z', fontsize=12)
ax.set_ylabel('Estimated z', fontsize=12)
ax.set_title('Starting Point z — True vs Estimated', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.set_xlim(0.15, 0.85)

plt.suptitle('Extended Parameter Recovery — True vs Estimated\n'
             '(v ∈ {0.1, 0.3, 0.5, 0.8}, z ∈ {0.3, 0.5, 0.7})',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('extended_scatter.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved.")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ── Left: spread of v estimates ──
ax = axes[0]
v_values = [0.1, 0.3, 0.5, 0.8]
for n in sample_sizes:
    subset = df_clean[df_clean['n_trials'] == n]
    for tv in v_values:
        data = subset[subset['true_v']==tv]['est_v']
        ax.scatter([n]*len(data), data,
                   alpha=0.08, s=6, color=colors[n])
        ax.scatter(n, data.mean(), s=80, color=colors[n],
                   zorder=5)

ax.axhline(0.1, color='grey', linestyle=':', alpha=0.5)
ax.axhline(0.3, color='grey', linestyle=':', alpha=0.5)
ax.axhline(0.5, color='grey', linestyle=':', alpha=0.5)
ax.axhline(0.8, color='grey', linestyle=':', alpha=0.5)
ax.set_xlabel('Trials per simulation', fontsize=12)
ax.set_ylabel('Estimated v', fontsize=12)
ax.set_title('Drift Rate v — Spread by Sample Size', fontsize=12)
ax.set_xticks(sample_sizes)
ax.grid(True, alpha=0.3)

# ── Right: spread of z estimates ──
ax = axes[1]
z_values = [0.3, 0.5, 0.7]
for n in sample_sizes:
    subset = df_clean[df_clean['n_trials'] == n]
    for tz in z_values:
        data = subset[subset['true_z']==tz]['est_z']
        ax.scatter([n]*len(data), data,
                   alpha=0.08, s=6, color=colors[n])
        ax.scatter(n, data.mean(), s=80, color=colors[n],
                   zorder=5)

for tz in z_values:
    ax.axhline(tz, color='grey', linestyle=':', alpha=0.5)
ax.set_xlabel('Trials per simulation', fontsize=12)
ax.set_ylabel('Estimated z', fontsize=12)
ax.set_title('Starting Point z — Spread by Sample Size', fontsize=12)
ax.set_xticks(sample_sizes)
ax.grid(True, alpha=0.3)

plt.suptitle('Extended Recovery — Estimate Spread',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('extended_spread.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved.")
```
