```python
import sys
sys.path.append(r'C:\Users\ASUS\Documents\DDM')

from ddm_core import simulate_ddm
import numpy as np
import matplotlib.pyplot as plt

# Test: with high positive drift, most responses should be upper boundary
data = simulate_ddm(v=1.0, a=1.2, z=0.5, t0=0.15, n_trials=500)
print(f"Proportion upper boundary: {data[:,1].mean():.2f}")  # expect ~0.85+
print(f"Mean RT: {data[:,0].mean():.3f}")                    # expect ~0.4-0.6s

plt.figure(figsize=(10, 4))

# Separate upper and lower boundary responses
upper = data[data[:,1]==1, 0]
lower = data[data[:,1]==0, 0]

plt.hist(upper, bins=50, alpha=0.7, color='steelblue', label='Upper boundary')
plt.hist(lower, bins=50, alpha=0.7, color='coral', label='Lower boundary')
plt.xlabel('Reaction Time (s)')
plt.ylabel('Count')
plt.title('Simulated RT Distribution — DDM')
plt.legend()
plt.show()

print(f"Upper boundary trials: {len(upper)} ({len(upper)/len(data)*100:.1f}%)")
print(f"Lower boundary trials: {len(lower)} ({len(lower)/len(data)*100:.1f}%)")
print(f"Mean RT upper: {upper.mean():.3f}s")
print(f"Mean RT lower: {lower.mean():.3f}s")
```
